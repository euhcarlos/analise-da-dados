## Type

```python
x = 10
type(x)
print('x', type(x))

nome = 'Carlos'
print('nome', type(nome))

num = '100'
print('num', type(num))
num_convert = int(num);
print('num_convert', type(num_convert))

num_float = float(num);
print('num_float', type(num_float))
```
