## Lista

```python
lista = [4,'Carlos', 8.5] # Pode adicionar qualquer tipo de valor a lista
```

```python
type(lista)
```

```python
lista.append('Python') # Adicionar um elemeto na lista
```

```python
print(lista)
print(lista[1])
```

## Tupla

```python
tupla = (4,'Carlos', 8.5) # Não pode ter modificação
```

```python
type(tupla)
```

```python
tupla.append('Python')
```

```python
print(tupla)
```

## Dicionário

```python
dicionario = {'nome':'Carlos', 'cidade':'São Paulo', 'idade':30} # Lista com chave valor
```

```python
print(dicionario)
```

```python
type(dicionario)
```

```python
print(dicionario['nome'])
```

```python
dicionario['ultimo_nome'] = 'Eduardo'
```

```python
print(dicionario)
```
