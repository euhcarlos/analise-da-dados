# Operadores Lógicos



*   And
*   Or
*   Not



## Tabela verdade

**And**

```python
print(10 > 2 and 4 < 6)
print(10 > 2 and 4 > 6)
print(10 < 2 and 4 < 6)
print(10 < 2 and 4 > 6)
```

**Or**

```python
print(10 > 2 or 4 < 6)
print(10 > 2 or 4 < 6)
print(10 < 2 or 4 > 6)
print(10 < 2 or 4 < 6)
```

**Not**

```python
not 10 > 2
```

```python
not 10 < 2
```

# Variávies
Variáveis estão alocando espaços na memória para guarda arquivos

```python
# Nome da variável é "x" que armazena o valor 10 em um espaço na memoria
x = 10
print(x) # Referencia utiliza o nome atribuido a variável pelo programa
```

## Regras de variáveos



*   Começar com letra (a-z, A-Z) ou sublinhado ( _ )
*   São Case Sensitive (Diferença entre letras maiúscula e minúscula)

Exemplo:
  A !=a

```python
## E percebido a diferença maiúscula e minúscula
valor = 10
Valor = 20
_valor = 15

print(valor)
print(Valor)
print(_valor)
```

## Boas práticas para o nome da variável

*   Usar nomes que definam o que ela faz no programa
*   Não usar nomes longos Exemplo de soma de 2 números

> Recomendado: **soma = 2 + 5, soma_total  = 2 + 5, somaTotal = 2 + 5**

> Não recomendado: **a = 2 + 5, soma_dois_números_inteiros = 2 + 5, somaDeDoisNumerosInteiros = 2 + 5**



```python
## Bom

soma = 2 + 5
soma_total = 2 + 5
somaTotal = 2 + 5

## Ruim

a = 2 + 5
soma_dois_números_inteiros = 2 + 5
somaDeDoisNumerosInteiros = 2 + 5
```
