# Condicional

```python
# if - se algo acontecer
if 5 < 10:
    print("Verdadeiro")
```

```python
# If/else - se algo acontecer, se não acontecer
num1 = 10
num2 = 20
if num1 < num2:
    print("O num1 é menor que num2")
else:
    print("O num2 é menor que o num1")
```

```python
# if / elif / else
num1 = 10
num2 = 20

if num1 < num2:
    print("O num1 é menor que num2")
elif num1 > num2:
    print("O num1 é maior que num2")
else:
    print("O num1 é igual ao num2")
```

## Condicionais com operadores lógicos

```python
num1 = 10
num2 = 20
num3 = 2

if (num1 > num2) and (num1 > num3):
    print("O num1 é o maior numero")
else:
    print("O num1 não é maior numero")
```
