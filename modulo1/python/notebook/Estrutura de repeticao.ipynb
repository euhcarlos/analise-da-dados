# For

## Executar comandos, enquanto percorre os itens de uma coleção
**Sintaxe:**
> for <variável> in <coleção>:
  > comando

**Formas de repetição**


*   Simples
*   Range
*   Enumerate



```python
# For range: Gera uma repetição de 0 a 9
for contador in range(10):
  print(contador)
```

```python
# For Range + break: Para o codigo quando condição for satisfeita
for contador in range(1000):
  print(contador)
  if contador == 5:
    break
```

```python
lista = [1,2,3,4,5]
```

```python
# For Simples: Percore os dados de uma lista
for val in lista:
  print(val)
```

```python
# For com Enumerate: mosta o valor e o indice
for i, val in enumerate(lista):
  print(i, val)
```

# For operadores compostos

### Operadores Compostos

*   Muito utilizado em laços de repetição
*   Código mais reduzido

    Operador        |||    Símbolo   |||    Exemplo   |||  Equivalente

*   Mais igual     |||      +=      |||    num += 1  |||  num = num +1     
*   Menos igual    |||      -=      |||    num -= 2  |||  num = num – 1       
*   Vezes igual    |||      *=      |||    num *= 5  |||  num = num * 5
*   Dividido igual |||      /=      |||    num /= 2  |||  num = num / 2
*   Módulo igual   |||      %=      |||    num %= 2  |||  num = num % 2




```python
lista2 = [2,3,4]
```

```python
for val in lista2:
  val += 1
  print(val)
```

```python
for val in lista2:
  val *= 2
  print(val)
```

## While
**Executar comandos enquanto uma condição é verdadeira**
**Sintaxe:**
> while <condicao>: comando
----
> while <condicao>: comando

> else:
  comando

```python
# While - executa ação enquanto verdadeiro
val = 2
while val < 10:
  print(val)
  val *= 2
```

```python
# While com else
val = 1
while val < 10:
  print(val)
  val *= 2
else:
  print('Fim')
# Executa quando a condição não for mais verdadeira
```
